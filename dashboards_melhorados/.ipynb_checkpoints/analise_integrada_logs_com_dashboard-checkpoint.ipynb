# 📊 Análise Integrada de Sessões com Detecção de Anomalias e Dashboard Interativo

```

import sqlite3
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```

# Conectar ao banco de dados e carregar os logs
con = sqlite3.connect("database.db")
logs = pd.read_sql_query("SELECT * FROM logs", con)
logs['data_hora'] = pd.to_datetime(logs['data_hora'])

# Criar colunas numéricas
logs['session_duration'] = logs.groupby("usuario")["data_hora"].transform(lambda x: (x.max() - x.min()).total_seconds())
logs['pages_visited'] = logs.groupby("usuario")["acao"].transform("count")
logs['session_duration'] = logs['session_duration'].fillna(0)
logs['pages_visited'] = logs['pages_visited'].fillna(1)
```

```

# Normalização
scaler = MinMaxScaler()
X = scaler.fit_transform(logs[['session_duration', 'pages_visited']])

# Modelo de detecção
model = IsolationForest(contamination=0.2, random_state=42)
logs['anomaly_score'] = model.fit_predict(X)
logs['classification'] = logs['anomaly_score'].apply(lambda x: 'Normal' if x == 1 else 'Suspeito')
```

```

# Visualização com Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=logs, x='session_duration', y='pages_visited', hue='classification', palette={"Normal": "blue", "Suspeito": "red"})
plt.title('Sessões Analisadas com Isolation Forest')
plt.xlabel('Duração da Sessão (s)')
plt.ylabel('Páginas Visitadas')
plt.grid(True)
plt.show()
```

```

# Opcional: Dashboard interativo com filtros (Gradio)
import gradio as gr

def filtrar_logs(usuario='', acao=''):
    df = logs.copy()
    if usuario:
        df = df[df['usuario'].str.contains(usuario, case=False, na=False)]
    if acao:
        df = df[df['acao'].str.contains(acao, case=False, na=False)]
    return df[['usuario', 'acao', 'data_hora', 'session_duration', 'pages_visited', 'classification']]

with gr.Blocks() as dashboard:
    gr.Markdown("### Dashboard Interativo - Sessões de Usuários")
    usuario_input = gr.Textbox(label="Filtrar por Usuário")
    acao_input = gr.Textbox(label="Filtrar por Ação")
    btn = gr.Button("Filtrar")
    saida = gr.Dataframe()
    btn.click(fn=filtrar_logs, inputs=[usuario_input, acao_input], outputs=saida)

dashboard.launch()
```
