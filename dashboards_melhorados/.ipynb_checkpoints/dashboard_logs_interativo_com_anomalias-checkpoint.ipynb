# 📈 Dashboard com Gráficos Interativos

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import MinMaxScaler
import sqlite3

# Conexão com banco de dados
con = sqlite3.connect("../database.db")
df = pd.read_sql_query("SELECT * FROM logs", con)

# Criar colunas numéricas fictícias a partir de 'data_hora' e 'acao'
df["data_hora"] = pd.to_datetime(df["data_hora"])

# Exemplo: duração da sessão por ID (assumindo que cada ID é uma sessão individual)
df["session_duration"] = df.groupby("usuario")["data_hora"].transform(lambda x: (x.max() - x.min()).total_seconds())

# Exemplo: contar quantas ações cada usuário fez por sessão
df["pages_visited"] = df.groupby("usuario")["acao"].transform("count")

# Preencher NaN com 0 (caso algum cálculo falhe)
df["session_duration"] = df["session_duration"].fillna(0)
df["pages_visited"] = df["pages_visited"].fillna(1)

# Normalização
scaler = MinMaxScaler()
x = scaler.fit_transform(df[["session_duration", "pages_visited"]])

# Isolation Forest
model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
df["anomalia"] = model.fit_predict(x)
df["anomalia"] = df["anomalia"].map({1: 0, -1: 1})

# Gráfico
plt.figure(figsize=(10,6))
plt.scatter(df["session_duration"], df["pages_visited"], c=df["anomalia"], cmap="coolwarm", edgecolor="k")
plt.xlabel("Duração da Sessão (s)")
plt.ylabel("Ações Registradas (pages_visited)")
plt.title("Detecção de Sessões Anômalas")
plt.grid(True)
plt.show()

# Mostrar anomalias
print(df[df["anomalia"] == 1].head(10))
```
